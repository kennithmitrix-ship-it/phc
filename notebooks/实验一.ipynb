# 独立工作提醒
- 我们仅在找到重新训练的时刻（测试样本上的误差最小的时期）后才更改学习率和批量大小。 否则，通过改进步骤，您可能会增加过拟合（模型在训练集上准确度提高，在测试集上面准确度下降）。

- 如果我们保持迭代次数不变，但增加批量大小，通常会稍微改善训练效果。这是因为当批量大小增加时，每个批次中包含更多的样本，这些样本提供了更多的信息来计算梯度。因此，我们可以更准确地估计整个数据集的梯度情况，从而使优化步骤更加精确。 但是由于各种原因，这个理论可能与您现在尝试做的内容结果出现不一致，理解理论就好

- 当调整学习率或批量大小时，应该相应地调整训练的周期数，以便保持总的训练迭代次数大致不变。例如，如果我们将学习率增加两倍，那么为了保持相似的训练迭代次数，可能需要将训练的周期数减半，以便在新的学习率下进行相似数量的参数更新。同样，如果我们将批量大小减少一半，也需要增加训练的周期数，以便在更小的批次上进行相似数量的参数更新。 在实践中，这种条件不是必需的，因为通过新的训练，许多超参数会立即更改。 但我们正在研究和探索不同超参数的行为，应该将每个超参数单独考虑，这意味着在比较不同学习率或批量大小时，应该尽可能保持其他超参数不变，以便更清晰地观察单个超参数对模型性能的影响。

- 当我们对模型进行修改或更换新的模型架构时，旧有的超参数可能不再适用或需要进行调整。 无需再次写出所有步骤，只需自己完成即可 - 重新训练并寻找更好的学习率和批量大小。

__创建模型__：

模型的创建是使用 nn 模块进行的，而最流行的神经网络块或层已经在该模块中实现，例如：
* Linear全连接线性层
* Conv2d卷积层
* MaxPool2d池化
* BatchNorm2d标准化
* 许多激活函数ReLU、Softmax、Tanh
* 正则化层，例如 Dropout

在本实验室工作中，我们将仅考虑上面列表中的 2 个神经网络构建块，即 Linear 和 ReLU。

您可以通过两种方式指定模型：

1. 使用nn.Sequential
2. 使用nn.Module类的继承

第一种方法适合创建没有分支的简单模型。 本质上，它们可以被认为是一条传送带，输入张量被传递到一系列顺序变换以产生输出张量。

当需要应用更复杂的架构，其中过程可以分叉成多个分支时，则使用 第二种方式 nn.Module。 这种方法允许您实现各种架构。

要创建一个具有一个隐藏层和一个非线性函数的简单多层感知器，根据第一种方法，编写以下代码就足够了：

    model = nn.Sequential(
      nn.Linear(input_dims, hidden_dims),
      nn.ReLU(),
      nn.Linear(hidden_dims, num_classes)
    )


要创建一个具有一个隐藏层和一个非线性函数的简单多层感知器，根据第二种方法，您需要创建一个类和一个模型作为此类的实例：

    class MLP(nn.Module):
        def __init__(self, input_dims, hidden_dims, num_classes,
                     *args, **kwargs):
            super(MLP, self).__init__()
            self.fc1 = Linear(input_dims, hidden_dims)
            self.fc2 = Linear(hidden_dims, num_classes)
        
        def forward(self, input):
             x = self.fc1(input)
             x = F.relu(x)
             x = self.fc2(x)
             return x
    
    model = MLP(input_dims, hidden_dims, num_classes)


同时，您也可以将 nn.Module 和 nn.Sequential 嵌套在其他模块中，这样可以创建非常复杂的模型架构。

__模型训练__：

在训练模型之前，您需要选择损失函数和优化器。 nn 模块中还提供了各种损失函数：
* __nn.MSELoss__ - 均方误差 (y_true-y_pred)**2
* __nn.BCEWithLogitsLoss__ - 二元分类问题的二元交叉熵
* __nn.CrossEntropyLoss__ - 多类分类问题的分类交叉熵

作为替代方案，您可以自己实现损失函数，例如 MSELoss：

    inputs, y = batch
    ...
    output = model(inputs)
    loss = ((output - y)**2).sum()
    ...

优化器包含在 __torch.optim__ 模块中。 目标函数优化器有很多，最经典的是随机梯度下降（Stochastic Gradient Descent）或简称SGD。 您必须将神经网络模型的参数传递给优化器并指定指定学习率（learning rate）等超参数。

要将模型置于训练状态，必须调用 __train__ 方法。 之后模型就可以进行训练了。

为了训练神经网络模型，使用基于逐次逼近方法的梯度下降及其变体。

在一次迭代中，我们有条件地选择将迭代器传递给整个数据集；在一次迭代中，我们选择使用当前的小批量来优化模型参数。 当对损失函数调用 __backward__ 方法时，PyTorch 会自动计算导数。

在这种情况下，当再次调用时，新的梯度值将被添加到之前计算的梯度值中。 因此，为了避免不必要的影响，通常的做法是使用应用于优化器类实例的 __zero_grad__ 方法在每次迭代时清除过去的梯度值。

为了更新神经网络的参数，使用 __step__ 方法，应用于优化器类的实例。

__模型质量检查__：

要将模型置于测试状态，您必须调用 __eval__ 方法。 之后模型就可以进行测试了。

模型预测的输出张量必须从计算图中切断。 为此，请使用应用于模型输出张量的 __detach__ 方法。 否则，可能会发生内存泄漏。 __numpy__ 方法将张量转换为多维 NumPy 数组。

默认情况下，模型输出所谓的类的逻辑值logits，而不是它们的概率。 为了获得概率，需要使用激活函数__Softmax__。 然而，实际上这不是必需的，因为logits的值与类别的概率一致，可以直接使用 logits 来获得最可能的类别。 使用 __argmax__ 或 __argsort__ 方法，可以获得最可能的类别的编号。argsort 方法还可以用于计算一些指标，比如 Accuracy@5 和排序指标。

# 第 1 部分. 使用万能逼近定理的回归问题，手动微分

## 生成样本并初始化神经网络参数

```python
X = (np.arange(100)/100 - 0.5).repeat(5)

y = 1/(1+np.exp(-10*X))
yn = np.random.normal(scale=0.05, size=y.size)+y

plt.plot(X, yn)
plt.plot(X, y, linestyle='--', c='k')
################################################
tensor_X = torch.Tensor(X.reshape(-1, 1))
tensor_y = torch.Tensor(yn.reshape(-1, 1))

HIDDEN_SIZE = 64
# 使用一个隐藏层初始化 MLP 权重
weights_1 = (torch.rand(1, HIDDEN_SIZE)-.5)/10
bias_1 = torch.zeros(HIDDEN_SIZE)

weights_2 = (torch.rand(HIDDEN_SIZE, 1)-.5)/10
bias_2 = torch.zeros(1)
```

## 训练回归任务的神经网络

```python
# 定义非线性函数
relu = lambda x: torch.maximum(x, torch.Tensor([0]))
# 前向传播
forward = lambda x: (weights_2.t()*relu((weights_1*x) + bias_1)
                      ).sum(axis=-1,keepdims=True) + bias_2
loss = lambda y, y_: ((y-y_)**2).sum(axis=-1)

# #反向传播
def backward(X, y, y_pred):
    # 损失函数相对于 y_pred 的导数
    dL = 2*(y_pred-y)
    # 应用激活之前的隐藏层神经元值
    Ax = (weights_1*X) + bias_1
    # 应用激活后的隐藏层神经元值
    A = relu(Ax)
    # 损失函数相对于weight_2的导数
    dW2 = torch.mm(A.t(), dL)
    # 损失函数相对于bias_2的导数
    db2 = dL.sum(axis=0)
    # 损失函数相对于激活后隐藏层值的导数
    dA = torch.mm(dL, weights_2.t())
    # 损失函数相对于激活前隐藏层值的导数
    dA[Ax<=0] = 0
    # 损失函数相对于weight_1的导数
    dW = torch.mm(X.t(), dA)
    # 损失函数相对于bias_1的导数
    db = dA.sum(axis=0)
    #print(dW.shape, db.shape, dW2.shape, db2.shape)
    return dW, db, dW2, db2

def optimize(params, grads, lr=0.001):
    # 整个训练集的梯度下降
    W1, b1, W2, b2 = params
    W1 -= lr*grads[0]
    W2 -= lr*grads[2]
    b1 -= lr*grads[1]
    b2 -= lr*grads[3]
    return W1, b1, W2, b2

for i in range(50000): # 梯度下降 5 万次迭代 == 5 万个 epoch
  output = forward(tensor_X)
  cur_loss = loss(output, tensor_y)
  grads = backward(tensor_X, tensor_y, output)
  params = [weights_1, bias_1, weights_2, bias_2]
  weights_1, bias_1, weights_2, bias_2 = optimize(params, grads, 1e-4)
  if (i+1)%10000 == 0:
      plt.plot(X, output.numpy(), label=str(i+1), alpha=0.5)
plt.plot(X, y, linestyle='--', c='k', label='real')
plt.legend()
plt.ylim(y.min(), y.max())
print(cur_loss.numpy().mean())
```

# 第 2 部分：使用 PyTorch 自微分进行二元分类

## 生成样本并初始化神经网络参数

```python
X = np.random.randint(2, size=(1000, 2))

y = (X[:, 0] + X[:, 1]) % 2 # XOR
X = X + np.random.normal(0, scale=0.1, size=X.shape)
#X, y = make_circles(n_samples=1000, noise=0.025)
#X, y = make_moons(n_samples=1000, noise=0.025)
plt.scatter(X[:, 0], X[:, 1], c=y)
####################################################
tensor_X = torch.Tensor(X.reshape(-1, 2))
tensor_y = torch.Tensor(y.reshape(-1, 1))

HIDDEN_SIZE = 16
# 使用一个隐藏层初始化 MLP 权重
weights_1 = ((torch.rand(2, HIDDEN_SIZE)-.5)/10).detach().requires_grad_(True)
bias_1 = torch.zeros(HIDDEN_SIZE, requires_grad=True)

weights_2 = ((torch.rand(HIDDEN_SIZE, 1)-.5)/10).detach().requires_grad_(True)
bias_2 = torch.zeros(1, requires_grad=True)
```

## 训练用于分类任务的神经网络

```python
# 定义非线性函数
def sigmoid(x):
    return 1/(1+torch.exp(-x))

# 前向传播
def forward(x):
    hidden = torch.mm(x, weights_1) + bias_1
    hidden_nonlin = sigmoid(hidden)
    output = (weights_2.t()*hidden_nonlin).sum(axis=-1,keepdims=True) + bias_2
    return sigmoid(output)

# Logloss打印损失值
def loss(y_true, y_pred):
    return -1*(y_true*torch.log(y_pred)+(1-y_true)*torch.log(1-y_pred)).sum()

# 设定学习步骤
lr = 1e-3
# 设置迭代次数
iters = 10000
params = [weights_1, bias_1, weights_2, bias_2]
losses = []
for i in range(iters):
    output = forward(tensor_X)
    lossval = loss(tensor_y, output)
    lossval.backward() # 这里 autograd 发挥作用
    for w in params:
        with torch.no_grad():
            w -= w.grad*lr # 更新权重
        w.grad.zero_() # 将梯度设置为零，这样它们就不会在迭代中累积
    losses.append(lossval.item())
# 通过迭代显示损失函数的历史记录
plt.plot(losses)
```

## 检查训练结果

## 创建具有一个隐藏层的多层感知器的 Pytorch 模型

```python
class Normalize(nn.Module):
    def __init__(self, mean, std):
        super(Normalize, self).__init__()
        self.mean = torch.tensor(mean)
        self.std = torch.tensor(std)

    def forward(self, input):
        x = input / 255.0
        x = x - self.mean
        x = x / self.std
        return torch.flatten(x, start_dim=1) # nhwc -> nm

class Cifar100_MLP(nn.Module):
    def __init__(self, hidden_size=32, classes=100):
        super(Cifar100_MLP, self).__init__()
        # https://blog.jovian.ai/image-classification-of-cifar100-dataset-using-pytorch-8b7145242df1
        self.norm = Normalize([0.5074,0.4867,0.4411],[0.2011,0.1987,0.2025])
        self.seq = nn.Sequential(
            nn.Linear(32*32*3, hidden_size),
            nn.ReLU(),
            nn.Linear(hidden_size, classes),
        )

    def forward(self, input):
        x = self.norm(input)
        return self.seq(x)

HIDDEN_SIZE = 10
model = Cifar100_MLP(hidden_size=HIDDEN_SIZE, classes=len(CLASSES))
model
```

```python
EPOCHS = 250
steps_per_epoch = len(dataloader['train'])
steps_per_epoch_val = len(dataloader['test'])
for epoch in range(EPOCHS):  # 多次浏览数据集
    running_loss = 0.0
    model.train()
    for i, batch in enumerate(dataloader['train'], 0):
        # 接收一个小批量；批处理是[inputs, labels]的二元素列表
        inputs, labels = batch

        # 清除上次迭代中过去的梯度
        optimizer.zero_grad()

        # 前向+后向传递+优化
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        #loss = F.cross_entropy(outputs, labels)
        loss.backward()
        optimizer.step()

        # 计算统计数据
        running_loss += loss.item()
    print(f'[{epoch + 1}, {i + 1:5d}] loss: {running_loss / steps_per_epoch:.3f}')
    running_loss = 0.0
    model.eval()
    with torch.no_grad(): # 禁用自动微分
        for i, data in enumerate(dataloader['test'], 0):
            inputs, labels = data

            outputs = model(inputs)
            loss = criterion(outputs, labels)
            running_loss += loss.item()
    print(f'[{epoch + 1}, {i + 1:5d}] val loss: {running_loss / steps_per_epoch_val:.3f}')
print('Обучение закончено')
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt

# 假设 Cifar100_MLP 已经定义
class Normalize(nn.Module):
    def __init__(self, mean, std):
        super(Normalize, self).__init__()
        self.mean = torch.tensor(mean)
        self.std = torch.tensor(std)

    def forward(self, input):
        x = input / 255.0
        x = x - self.mean
        x = x / self.std
        return torch.flatten(x, start_dim=1)

class Cifar100_MLP(nn.Module):
    def __init__(self, hidden_size=32, classes=100):
        super(Cifar100_MLP, self).__init__()
        self.norm = Normalize([0.5074, 0.4867, 0.4411], [0.2011, 0.1987, 0.2025])
        self.seq = nn.Sequential(
            nn.Linear(32*32*3, hidden_size),
            nn.ReLU(),
            nn.Linear(hidden_size, classes),
        )

    def forward(self, input):
        x = self.norm(input)
        return self.seq(x)

# 数据加载器、损失函数和优化器设置
# 使用已加载的 dataloader 对象
train_loader = dataloader['train']
val_loader = dataloader['test']

# 假设训练轮数
epochs = 250
train_losses = []
val_losses = []

# 模型和优化器
# Re-initializing model, optimizer and criterion for a fresh training run for plotting.
# HIDDEN_SIZE and CLASSES are assumed to be available from previous cells.
model = Cifar100_MLP(hidden_size=HIDDEN_SIZE, classes=len(CLASSES))
optimizer = optim.Adam(model.parameters(), lr=0.001) # Using Adam as defined in the original cell
criterion = nn.CrossEntropyLoss()

# 训练和验证过程
for epoch in range(epochs):
    model.train()
    running_train_loss = 0.0
    for inputs, labels in train_loader:  # 训练数据
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_train_loss += loss.item()

    avg_train_loss = running_train_loss / len(train_loader)
    train_losses.append(avg_train_loss)

    # 验证过程
    model.eval()
    running_val_loss = 0.0
    with torch.no_grad():
        for inputs, labels in val_loader:  # 验证数据
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            running_val_loss += loss.item()

    avg_val_loss = running_val_loss / len(val_loader)
    val_losses.append(avg_val_loss)

    # 每一轮打印一次损失
    print(f"Epoch {epoch+1}/{epochs}, Train Loss: {avg_train_loss:.4f}, Val Loss: {avg_val_loss:.4f}")

# 绘制训练和验证损失曲线
plt.plot(range(epochs), train_losses, label='Training Loss', color='blue')
plt.plot(range(epochs), val_losses, label='Validation Loss', color='orange')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss over Epochs')
plt.legend()
plt.show()
```

* __np.random.normal__ - 生成一个数组，其中填充了随机正态值，其标准差通过比例参数指定，且平均值等于平均参数。 数组中的元素数量由传递给 size 参数的数字或列表指定。

* __np.transpose__ - 重命名多维数组的轴。 为了处理图像，接受两种格式：NHWC 和 NCHW（N 是数组中的图像数量，C 是通道数量，H 是高度，W 是宽度）。 多维数组作为参数传递，以及具有新轴排列的列表或元组。 例如 `np.transpose([[0, 1, 2, 3]], (1,0))` 将创建一个 2D 列向量 [[[0], [1], [2], [3]] 。 请注意，轴从 0 开始。实际上，它用于将 NHWC 转换为 NCHW，反之亦然。 在第一种情况下，0轴N保持在其第一位置，第一轴H和第二轴W向右移动一个位置，并且3轴-C放置在第二位置。 那些。 我们得到以下排列：[0, 3, 1, 2]
* __np.isin__ - 类似于 SQL 运算符 IN，逐个元素检查集合中是否包含数组。 `np.isin([0, 2, 1], [2, 3])` 将返回 [False, True, False]

* __np.unique__ - SQL 中 SELECT DISTINCT 的类似物。 使用标准参数，返回包含唯一元素的一维子数组。 如果指定设置 __return_inverse__ 标志，则将返回一个数组，其中包含具有唯一元素的数组样本的数量。 本质上是进行标签编码
* __np.concatenate__ - 沿指定轴串联多维数组。 轴编号通过 __axis__ 参数指定。 例如，它可用于组合多个特征或多个数据集。 在图像上下文中，它可用于将多个图像垂直和水平地组合或粘合成一幅图像。 在声音的背景下 - 将两个音轨粘合在一起。

* __plt.show__ - 强制绘制图形，可用于在一个代码块中显示多个图形。
* __plt.legend__ - 显示先前指定的图表标签
* __plt.xlim__ - 限制从第一个参数到第二个参数的 x 坐标范围。 默认情况下，水平轴范围是根据所使用的数据自动选择的。 手动设置横轴数值范围，使用此方法
* __plt.ylim__ - 与 __plt.xlim__ 类似，但用于垂直轴。
